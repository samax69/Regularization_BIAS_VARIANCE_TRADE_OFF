```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_samples = 100
slope = 3
intercept = 5
noise_level = 40

# Generate input features
X = np.random.rand(n_samples, 1) * 100  # Random values from 0 to 100

# Generate linear target with noise
noise = np.random.randn(n_samples, 1) * noise_level
y = slope * X + intercept + noise

# Plot the dataset
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Dataset with Noise')
plt.show()
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2)
```

```python
class SimpleRidge:
    def __init__(self , alpha = 0.1):
        self.alpha = 0.1
        self.coef = None
        self.intercept = None
    
    def fit(self , X , y):
        num = 0
        den = 0
        for i in range(X.shape[0]):
            num += (y[i] - y.mean())*(X[i] - X.mean())
            den += (X[i] - X.mean())*(X[i] - X.mean())

        self.coef = num / (den + self.alpha)
        self.intercept = y.mean() - self.coef * X.mean()
        print("Estimate Coef = " , self.coef)
        print("Estimate intercept = " , self.intercept)

    def predict(self , X):
        return self.coef * X + self.intercept
    
        
```
